```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)



cars1 = pd.read_table('Dataframe_propre.csv', sep = ",")
cars1.head(5)
```

```python
colonne = ["horsepower"]
X = cars1.loc[:, colonne]

# 2. instantiate model
linreg = LinearRegression()

# 3. fit 
Y = cars1["price"]
linreg.fit(X, Y)
```

```python
b = linreg.intercept_
b
```

```python
a = linreg.coef_
a
```

```python
z = a*X+b

z
```

```python
prediction = linreg.predict(X)
prediction
```

```python
import seaborn as sns
sns.set(color_codes=True)
sns.residplot(X,Y, data=cars1, scatter_kws={"s": 120})
```
